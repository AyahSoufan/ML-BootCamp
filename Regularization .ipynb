```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

from scipy import stats
from scipy.stats import norm, skew #for some statistics
import statsmodels.api as sm
import statsmodels
from statsmodels.regression import linear_model
```

```python
#Load Data
train= pd.read_csv(r"C:\Users\Ayah\Desktop\bootcamp\Data\Credit.csv")
train.head()
```

```python
#train.Balance 
```

```python
Y = train['Balance']
X = train
del X['Balance']
train_dummy = pd.get_dummies(X)
train_num= train_dummy.select_dtypes([np.number])
train_num = sm.add_constant(train_num)
```

```python
model = sm.OLS(Y, train_num, missing='drop')
result=model.fit()
result.summary()
```

```python
results_Ridge = model.fit_regularized(L1_wt=0, alpha=0.4, start_params=result.params)
ridge = sm.regression.linear_model.OLSResults(model, 
                                              results_Ridge.params, 
                                              model.normalized_cov_params)
```

```python
ridge.summary()
```

```python
results_LASSO = model.fit_regularized(L1_wt=1, alpha=0.4, start_params=result.params)
LASSO = sm.regression.linear_model.OLSResults(model, 
                                              results_LASSO.params, 
                                              model.normalized_cov_params)
LASSO.summary()
```
