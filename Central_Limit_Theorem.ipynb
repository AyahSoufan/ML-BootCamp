```python
# Let X be the number of sequences of "HT" in 10 consecutive coin tosses. Simulate this 
# procedure 100 times and plot the distribution of X. (b) Now rerun this simulation 1000 times,
# and compute the average of X in each simulation (i.e. 1000 simulations of 100 instances of 10
# coin tosses. Whew!) Plot the distribution of the average of X across these simulations.
#(c) Make up your own random variable having to do with coin tosses and repeat exercises (a) and (b).
# You may find the numpy.random library to be helpful. To simulate a coin toss, you can use the
# command np.random.binomial(1, 0.5), and treat the outcome 1 as heads and 0 as tails.
import numpy as np
import seaborn as sns
```

```python
#Solution a

my_list=[]
X=[]

for j in range (100):
    my_list=[]
    for i in range(10):
        res=np.random.binomial(1, 0.5)
        my_list.append(res)
        #print(my_list)
    count=0
    for i in range (9):
        if (my_list[i] == 1) & (my_list[i+1]==0) :
            count+=1
    X.append(count)
    #print('X')
    #print(X)
```

```python
import pandas as pd
import numpy as np
import random
from matplotlib import pyplot as plt

ax = sns.distplot(X)
ax.set(xlabel='Binomial Distribution', ylabel='Frequency')
plt.title("Number of sequences of \"HT\" in 10 consecutive coin tosses with 1 sample")
plt.show()
```

```python
plt.hist(X, bins=5, normed=True, color='c', edgecolor='k', alpha=0.65)
plt.title('Number of sequences of "HT" (100) )')
plt.xlabel('Numbers')
plt.ylabel('count')
plt.show()
```

```python
#Solution a
import numpy as np
import statistics
X_ave=[]

for z in range (1000):
    X=[]
    for j in range (100):
        my_list=[]
        for i in range(10):
            res=np.random.binomial(1, 0.5)
            my_list.append(res)
            
        count=0
        for i in range (9):
            if (my_list[i] == 1) & (my_list[i+1]==0) :
                count+=1
        X.append(count)
    X_ave.append (statistics.mean(X) )
len(X_ave)   
```

```python
ax = sns.distplot(X_ave)
ax.set(xlabel='Binomial Distribution', ylabel='Frequency')
plt.title("Number of sequences of \"HT\" in 10 consecutive coin tosses with 1 sample")
plt.show()
```

```python
plt.hist(X_ave, bins =30, color='c', edgecolor='k', alpha=0.65)
plt.title('distribution of the average of X (1000) )')
plt.xlabel('mean')
plt.ylabel('count')
plt.axvline(statistics.mean(X) , color='k', linestyle='dashed', linewidth=1)
plt.text(statistics.mean(X) + statistics.mean(X)/10, 120 - 60,
         'Mean: {:.2f}'.format(statistics.mean(X)))
plt.show()
```
